# **Análisis de Clientes que NO Permancen en el Servicio (Churn)– TelecomX LATAM**


**1. Introducción:**

El presente informe tiene como propósito analizar el comportamiento de los clientes de TelecomX LATAM con respecto a su permanencia o deserción del servicio, comúnmente conocido como 'churn'. Este fenómeno, asociado a la pérdida de clientes, representa un desafío crítico para las empresas del rubro de telecomunicaciones, ya que implica no solo una disminución de ingresos, sino también un posible indicador de deficiencias en los procesos de captación, retención, experiencia del cliente y soporte postventa.

El objetivo general de este análisis es comprender qué factores influyen en la decisión de los clientes de dejar el servicio, segmentando la base de datos y explorando tanto patrones demográficos como conductuales. Además, se busca identificar correlaciones entre variables clave, con el fin de facilitar la toma de decisiones estratégicas orientadas a reducir la evasión y mejorar la experiencia general del cliente.


**2. Limpieza y Tratamiento de Datos:**

Se trabajó con un archivo JSON extraído desde una fuente externa. El proceso de limpieza y transformación incluyó:

*   Normalización de datos anidados con json_normalize.

*   Renombrado de columnas para mayor claridad y usabilidad.

*   Conversión de variables categóricas a binarios (Sí/No a 1/0) para servicios contratados.
*   Conversión de columnas como 'Cargo total' a tipo numérico (float).

*   Identificación y manejo de valores nulos, vacíos o mal codificados.

*   Agrupación de rangos para análisis por niveles de cargo mensual y total.

*   Creación de columnas derivadas como Nº de servicios para cuantificar los servicios contratados por cliente.

*   El dataset final contó con 7043 registros correctamente estructurados.


**3. Análisis Exploratorio de Datos:**

Durante la fase exploratoria se desarrollaron múltiples visualizaciones interactivas con plotly express para entender mejor el perfil de los clientes que no permanecen en el servicio, que permitieron detectar patrones importantes:

* Distribución general:
Gráfico de torta mostró proporción entre clientes que permanecen y los que no.

* Comparaciones por variables clave:
Género y Edad (≥65 años) no presentaron diferencias significativas, aunque los mayores se mostraron más fieles.

* Antigüedad fue un factor importante: los clientes con menos meses de contrato abandonan más.

* Tipo de contrato reveló que quienes tienen contrato "mes a mes" son mucho más propensos a irse.

* Cantidad de servicios contratados: a mayor cantidad de servicios, menor abandono (más compromiso).

* Tipo de servicio de Internet: los Clientes que NO Permancen en el Servicio fue más alto en clientes con “Fiber optic” que en “DSL”.

* Forma de pago: el uso de “Electronic Check” estuvo fuertemente asociado al abandono.

* Factura electrónica no mostró un efecto marcado.

* Cargo mensual y cargo total: clientes que pagan poco y tienen bajo cargo total abandonan más, lo que puede indicar una mala primera experiencia.


**4. Conclusiones e Insights:**

El análisis sugiere que existen múltiples factores relacionados con la evasión de clientes, entre los que destacan:

1. **Tipo de contrato**: Los clientes con contrato 'Mes a mes' tienen una tasa de deserción significativamente mayor, lo que indica la necesidad de fortalecer estrategias de fidelización para este segmento.

2. **Servicios contratados**: Existe una correlación inversa entre el número de servicios contratados y el churn. A mayor cantidad de servicios, menor es la tasa de abandono, lo que demuestra la importancia del cross-selling.

3. **Tipo de Internet**: Los usuarios de 'Fiber Optic' presentan una mayor proporción de abandono que los de 'DSL'. Esto podría vincularse a expectativas no cumplidas o deficiencias de servicio.

4. **Forma de pago**: Los métodos electrónicos presentan mayor deserción que medios tradicionales como 'Cheque por correo', lo que podría relacionarse con el perfil digital de clientes más exigentes o autónomos.

5. **Cargo mensual y total**: Clientes con cargos altos y baja antigüedad tienen mayores tasas de churn, lo cual puede ser síntoma de una percepción de bajo valor del servicio recibido por el costo asumido.


**5. Recomendaciones Estratégicas:**

1. **Fortalecer el proceso de bienvenida, educación y acompañamiento del cliente (onboarding)**: Se debe diseñar un proceso estructurado de recepción que garantice que los nuevos clientes comprendan las funcionalidades, beneficios y uso del servicio contratado desde el primer día.

2. **Mejorar el compromiso, conexión y uso activo del servicio (engagement)**: Incentivar el uso activo de servicios con campañas personalizadas, beneficios por antigüedad y programas de lealtad que fomenten la retención.

3. **Fomentar contratos de mayor duración**: Promover beneficios asociados a contratos de uno o dos años, como descuentos o servicios adicionales gratuitos.

4. **Segmentar campañas según perfil de riesgo**: Usar los patrones detectados para identificar clientes con mayor probabilidad de evasión y aplicar campañas específicas de retención anticipada, (clientes con cargo mensual bajo y poco nivel de compromiso).
